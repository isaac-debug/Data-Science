```python
df = pd.read_csv("Taipei Income.csv", encoding = "big5")
df = df.drop(index=[0,1,2])
df.columns = ["year", "平均每戶經常性收入", "平均每戶所得收入"]
df
```

```python
for i in df.index:
  for col in df.columns:
    if col == "year":
      df[col][i] = int(df[col][i].strip()[:3])
    else:
      df[col][i] = int(df[col][i])

df = df.set_index("year")
df = df.transpose()
TaipeiIncome = df
TaipeiIncome["county"] = "臺北市"
TaipeiIncome
```

```python
df = pd.read_csv("NTPC Income.csv", encoding = "big5")
df.columns = ["year", "平均每戶經常性收入", "平均每戶所得收入"]
df = df.drop(index=list(range(8,23))).drop(index=0)
df
```

```python
for i in df.index:
  for col in df.columns:
    if col == "year":
      df[col][i] = int(df[col][i]) - 1911
    else:
      df[col][i] = int(df[col][i].replace(",", ""))

df = df.set_index("year")
NtpcIncome = df.transpose()
NtpcIncome["county"] = "新北市"
NtpcIncome
```

```python
df = pd.read_csv("Yilan Income.csv", encoding = "big5")
df.columns = ["year", "平均每戶所得收入", "平均每戶經常性收入"]
df = df.drop(index=list(range(9,23))).drop(index=[0,1])
df
```

```python
for i in df.index:
  for col in df.columns:
    if col == "year":
      df[col][i] = int(df[col][i]) - 1911
    else:
      df[col][i] = int(df[col][i].replace(" ", ""))

df = df.set_index("year")
YilanIncome = df.transpose()
YilanIncome["county"] = "宜蘭縣"
YilanIncome
```

**整合**

```python
Income = pd.concat([TaipeiIncome, NtpcIncome, YilanIncome])
Income = Income.reset_index().set_index(["county", "index"])
Income.index.set_names(["county", "income"], inplace = True)
Income = Income.sort_index(level = ["county", "income"], ascending = [False, True])
Income
```

```python
IncomeGrowth = Income[[]]
for i in Income.columns:
  try:
    growth = (Income[i] - Income[i-1])/Income[i-1]
    IncomeGrowth[i] = list(growth)
  except:
    IncomeGrowth[i] = math.nan

IncomeGrowth
```

```python
IncomeWithGrowth_list = []

for i in Income.columns:
  for j,k in Income.index:
    if k == "平均每戶所得收入":
      d = {}
      d["year"] = i
      d["county"] = j
      d[k] = Income[i][j,k]
      d[k + " Growth"] = IncomeGrowth[i][j,k]
      IncomeWithGrowth_list.append(d)
    else:
      IncomeWithGrowth_list[-1][k] = Income[i][j,k]
      IncomeWithGrowth_list[-1][k + " Growth"] = IncomeGrowth[i][j,k]

IncomeWithGrowth = pd.DataFrame(IncomeWithGrowth_list)
IncomeWithGrowth = IncomeWithGrowth.set_index(["year", "county"])
IncomeWithGrowth
```
